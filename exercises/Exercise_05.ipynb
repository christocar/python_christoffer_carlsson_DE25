```python
# be användaren att ange basmåttet och höjden på en triangel

base = float(input("Ange basmåttet på triangeln: "))
height = float(input("Ange höjdenb på triangeln: "))

# beräkna arean på triangeln
area = 0.5 * base * height

print("Arean på triangeln är:", area)
```

```python
# beräkna den euklidska avståndet mellan två punkter

import math # för att använda kvadratrotsfunktionen

x1 = float(input("Ange x-koordinaten för punkt 1: "))
y1 = float(input("Ange y-koordinaten för punkt 1: "))
x2 = float(input("Ange x-koordinaten för punkt 2: "))
y2 = float(input("Ange y-koordinaten för punkt 2: "))

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2) # euklidska avståndet formel
print("Det euklidska avståndet mellan punkterna är:", distance) 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Skapa x-värden från -10 till 10
x = np.linspace(-10, 10, 400)

# definiera funktionerna
f = lambda x: x**2 - 3
g = lambda x: 4*x - 7

# beräkna y-värden för varje x
y_f = f(x)
y_g = g(x)

# skapa plottarna
plt.plot(x, y_f, label='f(x) = x^2 - 3')
plt.plot(x, y_g, label='g(x) = 4x - 7')

# lägg till etiketter och titel
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plott av funktionerna f(x) och g(x)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
def clean_string(text):
    # ta bort mellanslag i början och slutet
    text = text.strip()
    # konvertera till gemener
    text = text.lower()
    # konvertera första bokstaven till versal
    text = text.capitalize()
    return text

# testa funktionen
test_string = ["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "  ]

new_strings = [clean_string(s) for s in test_string]
print(new_strings)
```

```python
def swedish_currency_sorting(amount):
    bills_and_coins = [1000, 500, 200, 100, 50, 20, 10, 5, 1]
    result = {}
    for bill in bills_and_coins:
        count = amount // bill
        if count > 0:
            result[bill] = int(count)
            amount -= bill * count
    return result

# testa funktionen
amount = int(input("Ange ett belopp i SEK: "))
sorting = swedish_currency_sorting(amount)
print("Valörer för beloppet:")
for bill, count in sorting.items():
    print(f"{count} x {bill} SEK")
```
