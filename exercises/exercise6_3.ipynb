# plotter library

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 2 * x + 5 + np.random.normal(0, 2, size=len(x))

print(f"{x[:5] = }")
print(f"{x[-5:] = }")
print(f"{x.shape = }")

fix, ax = plt.subplots(1)

ax.plot(x, y)
ax.set(title="Cool stock", xlabel="Days", ylabel="SEK")
ax.spines[["top", "right"]].set_visible(False)
ax.annotate("Big dip here", xytext = (9, 10), xy=(7.5, 14.2), arrowprops=dict(arrowstyle="->"))
```

## Refactor into a class

```python
class PlotAnnotator:
    def __init__(self, fig, ax):
        self.fig = fig
        self.ax = ax

    def arrow(self, label: str, start_xy: tuple[float], end_xy: tuple[float]):
        self.ax.annotate(
            label,
            xytext=start_xy,
            xy=end_xy,
            arrowprops=dict(arrowstyle="->"),
        )

    def remove_clutter(self):
        self.ax.spines[["top", "right"]].set_visible(False)

fig, ax = plt.subplots(1)

ax.plot(x, y)

annotator = PlotAnnotator(fig, ax)
annotator.arrow(label="Big dip here", start_xy=(9, 10), end_xy=(7.5, 14.2))
annotator.arrow(label="Top here", start_xy=(7, 26), end_xy=(9, 25))
annotator.remove_clutter()
```
